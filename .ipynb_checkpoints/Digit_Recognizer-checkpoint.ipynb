### Reading the trainning datafile


````python
train_df = pd.read_csv('train.csv')

train_df.head()
````

### Viewing the distrubition of the observations that we have 


````python
ggplot(data = train_df) + geom_bar(aes(x = 'label', binwidth = 0.5)) +\
scale_x_continuous(breaks = range(0,10,1))

````

based on the previous graph, It appear a balanced number of observeration of about 4000 of ecah number. We will start by exploring a small of the data that is easy to process and understand in order to gain insight about the data


### Exploring the vlues of the pixels

**Getting A sample of the training data**


````python
# setting the seed to keep woriking with same random numbers

np.random.seed(2018)

sample_ids = np.random.randint(0, len(train_df),size=200)

sample_ids.sort()

sample_df = train_df.iloc[sample_ids]

sample_df.head()
````

**Converting the data into long format for visualisation**


````python
# adding a colum for the observation id
sample_df.index.values

sample_df['id_obs'] = sample_df.index.values

sample_df.columns
````

````python
# renaming the columns
sample_df.columns = ['label'] + range(0,784) + ['id_obs']

sample_df.columns
````

````python
sample_df_stack = sample_df.melt(id_vars=['id_obs', 'label'], value_vars=range(0,784), var_name='pixel' ,value_name='pixel_value')

sample_df_stack[sample_df_stack.id_obs == 423]
````

````python
sample_df_stack2 = sample_df_stack.copy()

sample_df_stack2.id_obs = ['obs'+str(id) for id in sample_df_stack2.loc[0:,'id_obs']]

sample_df_stack2.head()

ggplot(data=sample_df_stack2) +\
geom_tile(aes(x = 'pixel', y = 'id_obs', fill = 'pixel_value')) +\
scale_fill_gradient(low = 'white', high = "black") +\
scale_x_discrete(breaks=range(0, 784, 256)) +\
facet_wrap('~label', scales='free_y') 

````

````python
sample_df_mean = sample_df_stack2.groupby(['label', 'pixel'], as_index=False,).mean()

sample_df_mean['y'] = [28 - (p / 28) for p in sample_df_mean.pixel]
sample_df_mean['x'] = [p % 28 for p in sample_df_mean.pixel]

sample_df_mean.head()
````

````python
ggplot(data=sample_df_mean) +\
geom_tile(aes(x = 'x', y = 'y', fill = 'pixel_value')) +\
scale_fill_gradient(low = 'white', high = "black") +\
scale_x_continuous(breaks=range(0, 28, 7)) +\
facet_wrap('~label', scales='free_y') 
````

# Applying classificatin models
## 1. Decision Trees



````python
from sklearn import tree

sample_df.head()

X = sample_df.loc[:, range(0,784)]
Y = sample_df.loc[:,'label']

clf = tree.DecisionTreeClassifier() 
clf = clf.fit(X, Y)


````

````python
# getting the remaing trainning data
test_ids = [i for i in range(0, len(train_df)) if i not in sample_ids]

test_ids[0:10]

test_sample_df = train_df.iloc[test_ids[0:100]]

X_p = test_sample_df.loc[:, ['pixel' + str(p) for p in range(0,784)]]
Y_p = clf.predict(X_p)

Y_p[0:10]
````

````python
test_sample_df.loc[:,'label'][0:10]
````

````python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data, format="png")
graph.render("digit_recognizer")
````
